```python
# HardCoding Variables
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time):
  a = 0.1
  b = -1
  c = 30

  temperature = a *(time ** 2) + b * time + c
  return temperature

def main():
  time_values = np.linspace(0, 10, 50)
  temperature_hardcoded = quadratic_model(time_values)
  plt.plot(time_values, temperature_hardcoded, label='Hard-coded Coefficient')
  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.legend()
  plt.title('Weather Modelling with Quadratic Equation')
  plt.grid()
  plt.show()
if __name__ == "__main__":
  main()
```

```python
#Single set of input

import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, coefficients):
    coefficients_list = list(coefficients)
    a = coefficients_list[0]
    b = coefficients_list[1]
    c = coefficients_list[2]
    temperature = a * (time ** 2) + b * time + c
    return temperature

def read_coefficients():
    coefficients = set()
    a = float(input("Enter the coefficient a: "))
    b = float(input("Enter the coefficient b: "))
    c = float(input("Enter the coefficient c: "))
    coefficients.add(a)
    coefficients.add(b)
    coefficients.add(c)
    return coefficients

def main():
    coefficients = read_coefficients()
    time_values = np.linspace(0, 10, 50)
    temperature_values = quadratic_model(time_values, coefficients)
    plt.plot(time_values, temperature_values, label='Single set of input')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation')
    plt.grid()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
#MUltiple set
import numpy as np
import matplotlib.pyplot as plt

def quadratic_models(time, a, b, c):
    temp = a * (time ** 2) + b * time + c
    return temp

time_vals = np.linspace(0, 10, 50)

def main():
    sets_of_coefficients = [(0.2, 1.5, 35), (-0.1, 2.0, 25), (0.3, -2, 30)]
    # Plot each set of coefficients
    for i, (a, b, c) in enumerate(sets_of_coefficients):
        temperature = quadratic_models(time_vals, a, b, c)
        plt.plot(time_vals, temperature, label=f'Set {i+1}: a={a}, b={b}, c={c}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Multi Set Quadratic Models')
    plt.grid(True)
    plt.show()

main()
```

```python
#multiple set from a file
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def read_coefficients_from_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        coefficients = []
        for line in lines:
            coeffs = [float(x) for x in line.strip().split()]
            coefficients.append(coeffs)
    return coefficients

def main():
    coefficients_list = read_coefficients_from_file('/content/mset.txt')
    time_values = np.linspace(0, 10, 50)

    for i, coefficients in enumerate(coefficients_list):
        a, b, c = coefficients
        temperature_values = quadratic_model(time_values, a, b, c)
        plt.plot(time_values, temperature_values, label=f'Set {i+1}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```

```python
#Hardcoded Coefficients and user input
import numpy as np
import matplotlib.pyplot as plt

def quadratic_models(time, a, b, c):
    temp = a * (time ** 2) + b * time + c
    return temp

time_vals = np.linspace(0, 10, 50)

def main():
    # Hardcoded coefficients
    a, b, c = 0.1, -1, 30
    temp_hardcoded = quadratic_models(time_vals, a, b, c)
    plt.plot(time_vals, temp_hardcoded, label=f'Hardcoded coeff: a={a}, b={b}, c={c}')

    # User input coefficients
    a = float(input("Enter the value of a: "))
    b = float(input("Enter the value of b: "))
    c = float(input("Enter the value of c: "))
    temp_user_input = quadratic_models(time_vals, a, b, c)
    plt.plot(time_vals, temp_user_input, label=f'User Input: a={a}, b={b}, c={c}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.title("Hardcoded and Input Coefficients")
    plt.legend()
    plt.grid()
    plt.show()

main()
```
